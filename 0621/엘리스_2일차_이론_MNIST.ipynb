### 0. 라이브러리 임포트

```python
import tensorflow as tf
import numpy as np
```

### 1. 데이터셋 구축

- 데이터 불러오기

```python
mnist_data = tf.keras.datasets.mnist.load_data()
```

- 데이터 split 하기

```python
(x_train,y_train), (x_test,y_test) = mnist_data
```

```python
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

### 2. 모델 구축하기

- 모델의 뼈대를 잡는게 있었죠. 뭘로? tf.keras.Sequential(layer를 쌓았죠)

image.png

```python
model = tf.keras.models.Sequential([
    # input layer , 여기는 input shape을 지정해줘야합니다.
    # flatten을 통해 (28,28) => (784,)
    tf.keras.layers.Flatten(input_shape=(28,28)),
    # hidden layer 쌓기 , 1개만 쌓을게요. 여기서는 activation function을 지정해줘야함.
    tf.keras.layers.Dense(32, activation='sigmoid'),
    # output layer 쌓기 , 노드개수 = 라벨 갯수(0~9, 10개) (분류테스크이니까) , 분류니까 activation function은 softmax 여야함
    tf.keras.layers.Dense(10 , activation='softmax')
])
```

```python
model
```

- 모델의 옵션을 지정해주는 compile

```python
model.compile(
    optimizer = 'Adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['Accuracy']
)
```

### 3. 모델 학습하기

- model.fit 을 통해 학습

```python
model.fit(
    x_train,
    y_train,
    epochs=3,
    # batch_size = 100,
)
```

### 4. 평가 및 추론하기

```python
loss , accuracy = model.evaluate(
    x_test,
    y_test
)
```

```python
result = model.predict(
    x_test
)
```
